import libraries and a dataset for two different types of datasets

```python
import pandas as pd
import os

path_1=os.getcwd()+"\\data\\fire_archive_M-C61_626683.csv.xz"
path_2=os.getcwd()+"\\data\\fire_archive_SV-C2_628846.csv.xz"
data_1=pd.read_csv(path_1)
data_2=pd.read_csv(path_2)
```

convert acq_date and acq_time to a single datetime column for both datasets

```python
data_1['acq_time'] = pd.to_datetime(data_1['acq_date'] + ' ' + data_1['acq_time'].astype(str).str.zfill(4), format='%Y-%m-%d %H%M')
data_2['acq_time'] = pd.to_datetime(data_2['acq_date'] + ' ' + data_2['acq_time'].astype(str).str.zfill(4), format='%Y-%m-%d %H%M')
data_1.drop(['acq_date'], axis=1, inplace=True)
data_2.drop(['acq_date'], axis=1, inplace=True)
```

Make both datasets have the same start date

```python
data_1=data_1[data_1['acq_time'] >= data_2['acq_time'].min()].copy()
```

get binned confidence for the first dataset. The second is already binned.

```python
data_1['confidence']=pd.cut(data_1['confidence'],bins=[-1,30,80,101],labels=['l','n','h'])
```

use ks_2samp to check if the datasets are of the same population

```python
from scipy.stats import ks_2samp

data_1['acq_time'] = pd.to_datetime(data_1['acq_time']).dt.date.apply(lambda x: x.toordinal())
data_2['acq_time'] = pd.to_datetime(data_2['acq_time']).dt.date.apply(lambda x: x.toordinal())

categorical = ['satellite', 'daynight', 'version', 'type']
for cat in categorical:
    data_1[cat] = data_1[cat].map({name:i for i, name in enumerate(data_1[cat].unique())})
    data_2[cat] = data_2[cat].map({name:i for i, name in enumerate(data_2[cat].unique())})

stat, p = ks_2samp(data_1, data_2)
print(f"KS test p-value: {p}")
```
